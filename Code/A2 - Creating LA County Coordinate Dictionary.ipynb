# Creating the LA County Coordinate Dictionary

[nominatim.openstreetmap.org](https://nominatim.openstreetmap.org/) is an interface for searching the open-source map website [openstreetmap.org](https://openstreetmap.org/). OpenStreetMap has a rich library of information, including specific geographic coordinates for various areas.

## 1. Visit [nominatim.openstreetmap.org](https://nominatim.openstreetmap.org/) and search for Los Angeles County (or whichever locality you are interested in).

## 2. In the next screen, click on the `details` button in the top left.

## 3. Copy the six-digit number in the `OSM` field.

## 4. Visit [http://polygons.openstreetmap.fr/](http://polygons.openstreetmap.fr/) and paste the six-digit number into the `Id of relation` field and click `Submit`

## 5. Select and download the appropriate `.json` file

The strict boundaries are described by the `0` `params`, and increasing any of the other parameters adds to the padding.

Because we wanted to ensure that all points inside LA county were collected in the scrape, we set our boundaries to have a large amount of padding (0.02-0.01-0.01). Clicking on the `image` link will show how the padded boundary compares to the actual boundary.

If you are keeping our folder structure, be sure to download to the `../Assets` folder!
