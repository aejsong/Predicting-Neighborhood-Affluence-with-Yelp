```python
import pandas as pd
```

```python
df= pd.read_csv('../data/master.csv')
del df['Unnamed: 0']
df.head()
```

```python
zip_list = df['zip_code'].unique().astype(int)
```

```python
zip_list = zip_list.tolist()
```

```python
irs = pd.read_csv('../data/IRSIncomeByZipCode.csv')
irs.head(10)
```

```python
irs['ZIPCODE'].dtype
```

```python
zip_agi = irs.loc[irs['ZIPCODE'].isin(zip_list),['ZIPCODE','Avg AGI']]
```

```python
zip_agi.columns = ['zip_code', 'ave_agi']
```

```python
zip_agi
```

```python
df
```

```python
df_income = pd.merge(df, zip_agi,how = 'outer' )
df_income
```

```python
df_income.isnull().sum()
```

```python
df_income[df_income.isnull().any(axis=1)]
```

```python
df_income.dropna(inplace = True)
```

```python
df_income.isnull().sum()
```

```python
df_income['count'] = df_income.groupby('zip_code')['zip_code'].transform('count')
```
