# Collecting US Census Income Information

The [United States Census Bureau](https://www.census.gov/data.html) offers free access to all of its data via [data.census.gov](https://data.census.gov/). Searching for this data is not as intuitive as one would hope, however.

## 1. Visit [data.census.gov](https://data.census.gov/) and search for `S1901`.

A3_1.png

## 2. You want 2017 data, but it will take you to the 2018 data anyway. Navigate to "Product" under the title and select `2017: ACS 5-Year Estimates Subject Tables`

## 3. Click on the `Geographies` button.

## 4. Select `All ZCTAs in United States`

## 5. Download the file.
